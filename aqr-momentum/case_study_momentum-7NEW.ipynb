# Case Study Questions 

## Question 1 

*Should AQR launch the momentum fund?*

The answer depends on a combination of factors. First, timing is crucial. Abstracting for a moment from the long-only restriction, open-ended capital requirements, and transaction costs, early 2009 represented a market bottom. Empirically, momentum strategies, particularly 12-month momentum, tend to underperform during sharp market reversals. Compared to shorter-term momentum strategies (e.g., 3-month), 12-month momentum typically experiences significant underperformance in post-crisis recovery phases, with recovery periods that can extend beyond one year. Therefore, early 2009 would likely have been an unfavorable time to launch a momentum fund. Beyond timing, structural considerations related to the legal framework of mutual funds also matter. The long-only constraint presents both advantages and disadvantages, as illustrated in Exhibit 3. On the positive side, a long-only strategy involves fewer positions  (376 stocks on average versus 844 for the long-short portfolio), which is lower turnover and potentially reduced transaction costs, though boundary stock treatment is a serious consideration the  nature of long only  improves net alpha retention, as top performers typically persist over long periods. Additionally The long-only portfolio exhibits a higher average annual return (20.42%) compared to the long-short strategy (17.99%), implying an annual gap of approximately 2.42%. When combined with lower transaction costs, this makes the long-only structure potentially more cost-efficient for a mutual fund product, when compared to a long-short traditional momentum strategy. Furthermore, the estimated annual alpha is higher for the long-only strategy (4.44% versus 3.24%), and the minimum annual return is less severe (−46.34% versus −51.48%). However, the main issue lies in factor exposure. The long-only strategy exhibits a strong positive beta to the market (RMRF = 1.20), compared to -0.08 for the long-short strategy. This eliminates the market-neutral diversification benefit and implies that part of the higher return is driven by systematic market exposure rather than pure momentum alpha. Regarding the HML factor, a loading of -0.12 confirms the structural complementarity with value strategies, suggesting that combining long-only momentum with value could mitigate cyclicality. Another important issue concerns open-ended capital structures. A long-only strategy exhibits relatively high annual volatility  and the inability to lock in capital during periods of market stress can lead to investor redemptions precisely when the strategy is most vulnerable, amplifying drawdowns despite momentum doing best after falling out of favour or ‘failing’. This liquidity-risk interaction is particularly problematic for less conservative strategies such as momentum. Therefore, launching a long-only momentum fund in early 2009, without diversification through complementary strategies such as value, and given open-ended capital constraints, would likely not have been an optimal decision.

## Question 2 

*How important do you think it is that AQR establish a momentum index at the same time they launched their funds?*

Establishing a momentum index alongside the launch of the funds was a strategically sound decision. An index provides a transparent, rule-based benchmark that reinforces the structural integrity of the strategy and enhances its credibility with investors. In contexts where backtesting results cannot be used in marketing materials under U.S. securities regulations, the role of index build by a third part, becomes even more central, as it offers an objective and compliant reference framework for evaluating performance. The transparency can improve investor understanding and potentially increase patience during adverse periods, thereby mitigating redemption risk associated with open-ended capital structures. In this sense, the index also serves an educational role, clarifying that momentum is a systematic factor rather than a discretionary or opportunistic approach. Moreover, the existence of a dedicated momentum index reduces ambiguity in performance evaluation. Instead of being compared directly to the S&P 500, which may be inappropriate given the strategy’s factor exposure, the fund can be evaluated relative to a coherent factor benchmark. Even if the fund underperforms the index net of fees, such underperformance is conceptually different from underperforming a broad market index. The tracking error–outperformance trade-off becomes more transparent and intellectually defensible, as setting a defined maximum tracking error relative to a dedicated benchmark allows alpha retention to be realistically assessed, compared, and systematically justified. In conclusion, as previously noted with respect to U.S. securities regulations governing the marketing use of backtested or hypothetical performance, the rationale for launching an index extends beyond promotional considerations. Rather, the index enhances structural transparency, reinforces benchmark alignment, and improves expectation management by establishing a consistent and objective reference framework for performance evaluation.

Momentum strategies can be attractive products for mutual fund investors, particularly those who are less risk-averse and possess a higher degree of financial literacy. Investors must be willing to remain invested during such phases to allow the strategy to recover and fully express its risk premia. Traditionally, momentum has been more commonly associated with hedge fund strategies due to its dynamic nature and sensitivity to market regimes. However, informed mutual fund investors may still be attracted to systematic momentum exposure, especially if delivered in a transparent and disciplined framework. To address the limitations imposed by a long-only structure, a more appealing approach could involve building an adaptive framework that dynamically allocates across complementary quantitative strategies depending on the prevailing market phase. Empirically, momentum strategies tend to perform well during expansionary phases characterized by well-defined trends (Stage-2 phases). A 12-month momentum strategy, for instance, may identify long-term winners while short-term pullbacks (e.g., 1–2 month corrections) can provide more favorable entry points. A particularly robust implementation could involve a twin momentum framework, which combines 12-month momentum with short-term momentum signals to improve crash resilience. By blending slow-moving trend persistence with faster regime detection, such an approach may better navigate market downturns. Furthermore, integrating momentum with a value strategy, given their low or negative correlation, could create a more diversified and structurally resilient mutual fund product, provided the signals are properly constructed and combined. To conclude, a multi-factor structure, embedding momentum, value, and defensive factors would reduce cyclicality and improve the stability of returns across different market regimes resulting in a more stable and scalable investment product.

## Question 5

*How important do you think the constraint is that the funds are long-only and do not allow short positions? Is this likely to be more of a concern for momentum than other quantitative strategies?*

A long-only momentum strategy can still be effective. As shown in Exhibit 3, it performs strongly and is capable of generating significant alpha. However, by construction, it remains fully exposed to systematic market risk and, in particular, to momentum crash risk during sharp reversals. Without integration with regime detection mechanisms, volatility scaling, or complementary strategies, a long-only momentum approach can be highly vulnerable to mean-reversion phases. This makes it not only financially volatile but also psychologically demanding for investors who may struggle to endure prolonged drawdowns. Under a long-only constraint, momentum is likely to be more affected than more conservative strategies such as value. Momentum tends to suffer significantly during abrupt market reversals, especially when leadership rotates quickly. In contrast, value strategies, given their structural exposure to mean reversion, often perform relatively better during these phases. During strong expansionary phases with persistent trends (Stage-2 market phase), value may simply underperform the market rather than experience catastrophic losses. From a portfolio construction perspective, a strategy that underperforms during favorable market conditions but demonstrates resilience during adverse regimes may be preferable to one that delivers strong returns in expansions yet exposes investors to severe capital impairment in downturns. The asymmetry of losses matters more than the symmetry of gains: large drawdowns require disproportionately higher subsequent returns to recover, thereby compounding long-term performance drag. Moreover, strategies that protect capital during stress periods often improve the overall portfolio’s risk-adjusted profile by reducing volatility, drawdowns, and correlation spikes when diversification is most needed. In a multi-asset allocation context, resilience during crisis regimes can enhance compounding efficiency, stabilize investor behavior, and lower the probability of forced deleveraging or redemptions at unfavorable times. In conclusion, the capacity to preserve capital during adverse market cycles, thereby mitigating tail risk and maintaining structural robustness across regimes, may ultimately prove more valuable than maximizing returns in already favorable environments.

## Question 6

*If AQR launches the Momentum Funds, how should they weigh maximizing returns vs. minimizing tracking error? How should they manage the portfolio?*

If AQR were to launch its Momentum Funds, portfolio management should explicitly balance the trade-off between maximizing active returns and minimizing tracking error relative to the benchmark. Given that momentum is an active and implementation-sensitive strategy, the objective should be to retain as much net alpha as possible while keeping tracking error within a predefined and acceptable range. Because momentum strategies are inherently turnover-intensive and require periodic rebalancing, they cannot be implemented as purely passive market-replication strategies, even when constructed using transparent and rule-based methodologies. As a result, cost control becomes a central consideration in their design and implementation, and each rebalancing decision should be evaluated on a rigorous cost–benefit basis. For example, when a new stock enters the top quantile of the momentum ranking, the manager should assess whether the expected incremental alpha justifies the transaction costs associated with selling the previous holding and initiating the new position. Trades where expected excess returns are likely to be fully offset by transaction costs should be avoided. The optimal point lies where marginal expected alpha equals marginal implementation cost. Furthermore, concentrating on securities with sufficient market capitalization and liquidity reduces transaction costs by limiting price impact and compressing bid–ask spreads. Capacity discipline is therefore essential to preserving net performance, as a strategy’s scalability is structurally constrained by market depth, turnover intensity, and execution frictions. As assets under management expand, trade sizes grow relative to available liquidity, increasing market impact, widening effective spreads, and generating slippage. These costs can materially erode gross alpha and, if unmanaged, deteriorate the strategy’s risk–return profile. Preserving performance thus requires aligning fund size with the liquidity capacity of the investable universe, continuously monitoring trading intensity, and, when necessary, restricting inflows to prevent performance dilution. In this context, execution timing is equally important. Portfolio management should incorporate optimal execution principles, particularly the trade-off between size, liquidity, and trading horizon. Liquidity can be proxied by market depth and the ability of the market to absorb orders without generating significant price impact. Algorithmic execution strategies, such as slicing orders over time and concentrating execution during high-volume sessions, would reduce market impact, minimize implementation shortfall, and help control tracking error. From an overall perspective, the manager should aim to limit unnecessary rebalancing while maintaining a maximum acceptable tracking error threshold. The process can be structured in two steps: first, evaluate the net alpha of portfolio rotation after transaction costs; second, optimize execution to minimize impact while preserving alignment with the benchmark. This disciplined framework would improve net performance by maximizing alpha retention and controlling tracking risk.

```python
#| label: tbl-summary
#| tbl-cap: "Summary Statistics"

factors = ff[["mktrf", "smb", "hml", "mom"]].dropna()
factors.index = pd.to_datetime(factors.index)

mean_monthly = factors.mean()
std_monthly  = factors.std()
n_months     = len(factors)

# Average Annual Return (expected)
avg_annual = mean_monthly * 12

# Annualised volatility
std_annual = std_monthly * np.sqrt(12)

# Sharpe
sharpe = avg_annual / std_annual

# t-stat (monthly mean)
t_stat = mean_monthly / (std_monthly / np.sqrt(n_months))

annual = (1 + factors).resample("YE").prod() - 1
annual.index = annual.index.year

max_annual = annual.max()
max_year   = annual.idxmax()

min_annual = annual.min()
min_year   = annual.idxmin()

pct_positive = (annual > 0).sum() / len(annual)

summary = pd.DataFrame({
    "Average Annual Return (%)": (avg_annual * 100).round(2),
    "T-stat": t_stat.round(2),
    "Std Deviation of Annual Return (%)": (std_annual * 100).round(2),
    "Annualized Sharpe": sharpe.round(2),
    "Max Annual Return (%)": (max_annual * 100).round(2),
    "Max Annual Return Year": max_year,
    "Min Annual Return (%)": (min_annual * 100).round(2),
    "Min Annual Return Year": min_year,
    "% Positive Years": (pct_positive * 100).round(2)
}).T

print("\nSummary Statistics\n")
print(tab.tabulate(summary, headers="keys", tablefmt="github"))
```

The summary statistics indicate that the MOM factor exhibits both strong economic and statistical significance relative to the other Fama–French factors. With an average annual return of 7.48% and a t-statistic of 4.52, the historical mean is clearly distinguishable from zero under standard inference assumptions, suggesting a robust long-run premium. Its annualized Sharpe ratio of 0.46 is the highest among the four factors, slightly exceeding the market factor (0.44), implying comparatively attractive risk-adjusted performance. At the same time, MOM displays substantial volatility (16.29%) and pronounced tail risk, as reflected in its extreme minimum annual return of –64.28% in 1932, highlighting its exposure to sharp reversal episodes. Nevertheless, the factor delivered positive returns in approximately 75% of calendar years, indicating a relatively high frequency of success despite episodic crashes. Overall, the statistics suggest that momentum has historically provided a statistically significant and economically meaningful return premium, albeit accompanied by meaningful downside risk in extreme market environments.

```python
#| label: tbl-rolling
#| tbl-cap: "Rolling 10-Year Summary Statistics"

factor_mom = ff[["mom"]].dropna()
factor_mom.index = pd.to_datetime(factor_mom.index)

WINDOW = 120  # 10 years monthly

# Rolling cumulative return
roll_10y_cum = (1 + factor_mom).rolling(WINDOW).apply(np.prod, raw=True) - 1

# Rolling CAGR
roll_10y_cagr = (1 + roll_10y_cum) ** (1/10) - 1

# Drop initial NaNs
roll_10y_cum = roll_10y_cum.dropna()
roll_10y_cagr = roll_10y_cagr.dropna()

# Helper: decade labels
start_dates = roll_10y_cagr.index - pd.DateOffset(months=WINDOW - 1)
end_dates = roll_10y_cagr.index
decade_labels = pd.Series(
    [f"{s.year}-{e.year}" for s, e in zip(start_dates, end_dates)],
    index=roll_10y_cagr.index
)

# Compute stats
median_10y_cum = roll_10y_cum.median()
median_10y_cagr = roll_10y_cagr.median()
pct_positive_10y = (roll_10y_cum > 0).mean()

worst_10y_cagr = roll_10y_cagr.min()
worst_10y_end = roll_10y_cagr.idxmin()
worst_10y_decade = worst_10y_end.map(decade_labels)

best_10y_cagr = roll_10y_cagr.max()
best_10y_end = roll_10y_cagr.idxmax()
best_10y_decade = best_10y_end.map(decade_labels)

# Build summary table
rolling_summary = pd.DataFrame({
    "% Positive 10y Periods": (pct_positive_10y * 100).round(2),
    "Median 10y Cumulative Return (%)": (median_10y_cum * 100).round(2),
    "Median 10y CAGR (%)": (median_10y_cagr * 100).round(2),
    "Worst 10y CAGR (%)": (worst_10y_cagr * 100).round(2),
    "Worst 10y Decade": worst_10y_decade,
    "Best 10y CAGR (%)": (best_10y_cagr * 100).round(2),
    "Best 10y Decade": best_10y_decade
}).T

print("\nRolling 10-Year Summary Statistics\n")
print(tab.tabulate(rolling_summary, headers="keys", tablefmt="github"))
```


The rolling 10-year analysis of the MOM factor provides strong distributional evidence regarding its long-run performance. Historically, approximately 79.6% of all rolling decades delivered positive compounded returns, indicating that an investor entering the strategy at a random point in time had a high probability of achieving a positive outcome over a ten-year horizon. The median 10-year cumulative return of 132.9% corresponds to a median annualized growth rate (CAGR) of 8.82%, suggesting that a typical decade generated economically meaningful long-term growth. While the worst observed decade (1932–1942) produced a negative CAGR of –12.65%, highlighting substantial tail risk during extreme macroeconomic regimes, the best decade (1990–2000) achieved a CAGR of 16.39%, reflecting the strong performance of momentum during sustained trend environments. Overall, the historical distribution of rolling decade outcomes supports the view that the expected return of the MOM factor over the next decade is more likely than not to be significantly greater than zero, albeit with non-negligible downside risk in adverse regimes.

### Constructing the Industry Momentum Index
The Industry Momentum (IND_MOM) index is constructed following a standard cross-sectional momentum framework adapted to the industry level.

The assumption beyond the construction of the momentum industry:

- At each formation date, industries are ranked based on their cumulative returns over the t–12 to t–2 window. This specification excludes the most recent month (t–1) to mitigate short-term reversal effects and microstructure noise, in line with established momentum methodology. The cumulative return over this 11-month lookback window serves as the ranking signal.

- Industries are sorted cross-sectionally according to their formation-period performance. The top 33% of industries (winners) are assigned to the long portfolio, while the bottom 33% (losers) are assigned to the short portfolio. Positions are equally weighted within each side to avoid concentration bias and to maintain a pure cross-sectional exposure.

- The portfolio is rebalanced on a quarterly basis (every three months). At each rebalance date, the ranking procedure is repeated and portfolio constituents are updated accordingly. Positions are held constant between rebalancing dates, resulting in a non-overlapping three-month holding structure.

- The Industry Momentum index (IND_MOM) is computed as the equally weighted return of the long portfolio minus the equally weighted return of the short portfolio. For simplicity, transaction costs and implementation frictions are not incorporated, and the resulting long–short return series is treated as a theoretical index proxy for industry-level momentum exposure.

### Momentum Industry Index on the whole time series

```python
#| label: tbl-ind-mom-ret-full
#| tbl-cap: "Industry Momentum Strategy Returns over Full Sample"

ind_full_q3 = ind / 100  # Convert to decimals if in %

# 1) Formation signal: (t-12 to t-2) compounded
formation = (1 + ind_full_q3.shift(2)).rolling(window=11, min_periods=11).apply(np.prod, raw=True) - 1


# --- parameters ---
hold = 3  # 3-month holding / rebalance frequency

# 3) Define rebalance dates (every 3 months) AFTER formation is available
valid_dates = formation.dropna(how="all").index
rebalance_dates = valid_dates[::hold]  # non-overlapping quarterly schedule

n = formation.shape[1]
k = int(np.floor(n / 3))

# 4) Build strategy monthly return series by "freezing" masks for 3 months
ind_mom_q = pd.Series(index=formation.index, dtype=float)
long_q = pd.Series(index=formation.index, dtype=float)
short_q = pd.Series(index=formation.index, dtype=float)

for t in rebalance_dates:
    # ranks/masks computed ONLY at rebalance date t
    ranks_t = formation.loc[t].rank(ascending=False, method="first")
    long_mask_t  = ranks_t <= k
    short_mask_t = ranks_t > (n - k)

    # apply the SAME masks to the next 3 months of returns
    for h in range(1, hold + 1):
        dt = t + pd.offsets.MonthEnd(h)  # assumes MonthEnd index; adjust if needed
        if dt not in ind_full_q3.index:
            continue

        r = ind_full_q3.loc[dt]

        long_r = r.where(long_mask_t).mean()
        losers_r = r.where(short_mask_t).mean()
        short_r = -losers_r

        ind_mom_q.loc[dt] = long_r + short_r
        long_q.loc[dt] = long_r
        short_q.loc[dt] = short_r

out_full = pd.DataFrame({
    "IND_MOM": ind_mom_q,
    "LONG": long_q,
    "SHORT": short_q
}).dropna()
```

#### Main Statistics for Industry Momentum Index of Whole Series

```python
#| label: tbl-ind-mom-full
#| tbl-cap: "Statistics of Industry Momentum over Full Sample"

series_dict = {
    "IND_MOM": out_full["IND_MOM"],
    "LONG": out_full["LONG"],
    "SHORT": out_full["SHORT"],
}

def perf_summary_monthly(x: pd.Series, decade_months: int = 120) -> pd.Series:
    x = x.dropna().sort_index()
    x.index = pd.to_datetime(x.index)

    # ---- core moments (monthly) ----
    mean_m = x.mean()
    std_m  = x.std(ddof=1)
    n      = x.shape[0]

    # Expected annualised (arithmetic)
    avg_a  = mean_m * 12
    vol_a  = std_m * np.sqrt(12)
    sharpe = avg_a / vol_a
    tstat  = mean_m / (std_m / np.sqrt(n))

    # ---- Realized compounding over ENTIRE SAMPLE (professional) ----
    total_cum = (1 + x).prod() - 1

    # sample length in years (use months/12 because your data are monthly)
    n_years_sample = n / 12
    cagr_full = (1 + total_cum) ** (1 / n_years_sample) - 1 if n_years_sample > 0 else np.nan

    # ---- Realized annual returns (calendar years) ----
    ann = (1 + x).resample("YE").prod() - 1
    ann.index = ann.index.year

    max_ann = ann.max()
    max_yr  = ann.idxmax()
    min_ann = ann.min()
    min_yr  = ann.idxmin()
    pct_pos = (ann > 0).mean()

    # ---- 10-year rolling cumulative + CAGR ----
    roll_10y_cum = (1 + x).rolling(decade_months).apply(np.prod, raw=True) - 1
    roll_10y_cagr = (1 + roll_10y_cum) ** (12 / decade_months) - 1  # = 1/10 when 120 months

    med_10y_cum  = roll_10y_cum.median()
    med_10y_cagr = roll_10y_cagr.median()

    if roll_10y_cagr.dropna().empty:
        worst_10y_cagr, worst_label = np.nan, np.nan
        best_10y_cagr,  best_label  = np.nan, np.nan
    else:
        worst_end = roll_10y_cagr.idxmin()
        best_end  = roll_10y_cagr.idxmax()

        worst_10y_cagr = roll_10y_cagr.loc[worst_end]
        best_10y_cagr  = roll_10y_cagr.loc[best_end]

        worst_start = x.index[x.index.get_loc(worst_end) - (decade_months - 1)]
        best_start  = x.index[x.index.get_loc(best_end)  - (decade_months - 1)]

        worst_label = f"{worst_start.strftime('%Y-%m')} → {worst_end.strftime('%Y-%m')}"
        best_label  = f"{best_start.strftime('%Y-%m')} → {best_end.strftime('%Y-%m')}"

    return pd.Series({
        # Expected / arithmetic annualisation (your original block)
        "Average Annual Return (Arithmetic, %)": round(avg_a * 100, 2),
        "T-stat": round(tstat, 2),
        "Annual Volatility (%)": round(vol_a * 100, 2),
        "Annualized Sharpe (Arithmetic)": round(sharpe, 2),

        # Realized / compounded over full sample
        "Full-Sample Cumulative Return (%)": round(total_cum * 100, 2),
        "Full-Sample CAGR (Geometric, %)": round(cagr_full * 100, 2),

        # Annual calendar-year extremes
        "Max Annual Return (%)": round(max_ann * 100, 2),
        "Max Annual Return Year": int(max_yr) if pd.notna(max_yr) else np.nan,
        "Min Annual Return (%)": round(min_ann * 100, 2),
        "Min Annual Return Year": int(min_yr) if pd.notna(min_yr) else np.nan,
        "% Positive Years": round(pct_pos * 100, 2),

        # Rolling decade distribution
        "Median 10y Cumulative Return (%)": round(med_10y_cum * 100, 2) if pd.notna(med_10y_cum) else np.nan,
        "Median 10y CAGR (%)": round(med_10y_cagr * 100, 2) if pd.notna(med_10y_cagr) else np.nan,
        "Worst 10y CAGR (%)": round(worst_10y_cagr * 100, 2) if pd.notna(worst_10y_cagr) else np.nan,
        "Worst 10y Decade": worst_label,
        "Best 10y CAGR (%)": round(best_10y_cagr * 100, 2) if pd.notna(best_10y_cagr) else np.nan,
        "Best 10y Decade": best_label,

        # Counts
        "N months": int(n),
        "N years (calendar)": int(ann.shape[0]),
        "N years (sample/12)": round(n_years_sample, 2),
    })

summary_all = pd.DataFrame({name: perf_summary_monthly(s) for name, s in series_dict.items()})
summary_all

```

The table shows that the full-sample Industry Momentum (IND_MOM) strategy delivers modest performance overall, with an average annual return of 5.96%, a Sharpe ratio of 0.51, and a CAGR of 5.40%, indicating statistically significant but economically moderate risk-adjusted returns (t-stat 2.97). The key insight, however, lies in the decomposition of the long and short legs. The long portfolio performs strongly, generating a 15.89% average annual return, a 15.15% CAGR, and a solid Sharpe ratio of 0.87, with 75.93% positive years and a median 10-year CAGR of 17.32%, confirming that industry winners exhibit persistent continuation effects. In contrast, the short leg is structurally weak and highly unstable: it shows a −11.19% CAGR, a negative Sharpe ratio (−0.51), only 29.63% positive years, and severe long-horizon drawdowns, including a worst 10-year CAGR of −15.81%. This asymmetry suggests that the strategy’s performance is almost entirely driven by the long exposure, while the short side introduces substantial volatility and crash risk without delivering consistent negative return. Overall, industry momentum does not appear to be a particularly strong standalone long–short factor; its economic value lies primarily in the long continuation component rather than in a symmetric long–short premium.

Overall, the estimates suggest that while industry momentum is strongly related to stock level momentum, it may also contain a sectoral component linked to macro sensitivities, information spillovers, and capital reallocation across industries.

```python
#| label: tbl-ew-portfolio-full
#| tbl-cap: "Summary Statistics of an Equally Weighted Portfolio of Industry Momentum, Value, and Stock Momentum over the Full Sample"

# -------------------------------------------------------
# 1) Extract Value and Stock Momentum from FF
# -------------------------------------------------------
ff_subset = ff[["hml", "mom"]].dropna().copy()
ff_subset.index = pd.to_datetime(ff_subset.index)

# Convert to decimals if needed (FF often in %)
if ff_subset.abs().median().mean() > 0.5:
    ff_subset = ff_subset / 100.0

ff_subset.columns = ["VALUE", "STOCK_MOM"]

# -------------------------------------------------------
# 2) Industry Momentum (full sample)
# -------------------------------------------------------
IND_MOM_FULL = out_full["IND_MOM"].dropna().copy()
IND_MOM_FULL.index = pd.to_datetime(IND_MOM_FULL.index)
IND_MOM_FULL.name = "IND_MOM"

# -------------------------------------------------------
# 3) Align on common dates
# -------------------------------------------------------
common_idx = IND_MOM_FULL.index.intersection(ff_subset.index)

combined_data = pd.concat(
    [IND_MOM_FULL.loc[common_idx], ff_subset.loc[common_idx]],
    axis=1
)

# -------------------------------------------------------
# 4) Equally weighted portfolio (1/3 each)
# -------------------------------------------------------
EW_PORT = combined_data[["IND_MOM", "VALUE", "STOCK_MOM"]].mean(axis=1).dropna()

# -------------------------------------------------------
# 5) Summary statistics (ONLY for EW portfolio)
# -------------------------------------------------------
def perf_summary_monthly(x: pd.Series) -> pd.Series:
    x = x.dropna().sort_index()
    x.index = pd.to_datetime(x.index)

    mean_m = x.mean()
    std_m  = x.std(ddof=1)
    n      = x.shape[0]

    avg_a  = mean_m * 12
    vol_a  = std_m * np.sqrt(12)
    sharpe = avg_a / vol_a if vol_a != 0 else np.nan
    tstat  = mean_m / (std_m / np.sqrt(n)) if std_m != 0 else np.nan

    total_cum = (1 + x).prod() - 1
    n_years_sample = n / 12
    cagr_full = (1 + total_cum) ** (1 / n_years_sample) - 1 if n_years_sample > 0 else np.nan

    ann = (1 + x).resample("YE").prod() - 1
    ann.index = ann.index.year

    max_ann = ann.max()
    max_yr  = ann.idxmax()
    min_ann = ann.min()
    min_yr  = ann.idxmin()
    pct_pos = (ann > 0).mean()

    return pd.Series({
        "Average Annual Return (%)": round(avg_a * 100, 2),
        "T-stat": round(tstat, 2),
        "Annual Volatility (%)": round(vol_a * 100, 2),
        "Annualized Sharpe": round(sharpe, 2),

        "Full-Sample Cumulative Return (%)": round(total_cum * 100, 2),
        "Full-Sample CAGR (Geometric, %)": round(cagr_full * 100, 2),

        "Max Annual Return (%)": round(max_ann * 100, 2),
        "Max Annual Return Year": int(max_yr) if pd.notna(max_yr) else np.nan,
        "Min Annual Return (%)": round(min_ann * 100, 2),
        "Min Annual Return Year": int(min_yr) if pd.notna(min_yr) else np.nan,
        "% Positive Years": round(pct_pos * 100, 2),
        "N months": int(n),
    })

summary_EW_only = perf_summary_monthly(EW_PORT).to_frame(name="EW_PORT")

print("\nEqual-Weighted Portfolio — Full Sample Summary Statistics\n")
print(tab.tabulate(summary_EW_only, headers="keys", tablefmt="github"))
```

#### Index Construction
The index is a monthly, equally weighted composite designed to capture three related but conceptually distinct sources of cross sectional return variation. It combines industry momentum, which reflects persistence in relative performance across industries, with a value component proxied by the value factor, and a stock level momentum component proxied by the broad momentum factor. Each month, the three series are aligned on a common sample and combined with equal weights, so the index return represents the average performance of industry momentum, value, and stock momentum in that month. Performance is then assessed using annualised mean return, annualised volatility, a Sharpe ratio based on these annualised moments, and full sample compounded outcomes over the 413 month period.

#### Conclusion
Launching a combined Value, Momentum and Industry Momentum fund is a viable strategic move for AQR. The equal weighted composite demonstrates a robust historical Sharpe ratio of 0.64 with an 8.47 per cent volatility, outperforming standalone momentum. This efficacy is underpinned by the significant negative correlation between Value and Momentum, approximately negative 0.41, which serves as a structural hedge against momentum crashes. Industry momentum adds a macro dimension by capturing sector level trends often led by large cap stock momentum, providing a more comprehensive factor profile during persistent market trends.

A critical issue AQR must consider is the potential for factor redundancy. The high correlation between stock level and industry momentum suggests that the industry component may not provide a distinct risk premium. Instead, it might simply increase the fund's beta to the standard momentum factor. Without rigorous attribution, the apparent diversification benefits may be illusory, masking a concentrated exposure to a single factor family and failing to provide genuine incremental alpha.

Furthermore, equal capital allocation does not ensure equal risk contribution. Since momentum factors are typically more volatile than value, the portfolio’s risk profile may be disproportionately dominated by momentum. AQR should evaluate volatility aware scaling or a risk parity framework to maintain the defensive properties of the value factor. The composite’s 12.56 per cent drawdown in 2023 highlights that the inclusion of value does not entirely immunise the portfolio against sharp, synchronised momentum reversals.

Operational challenges also require careful management, particularly the higher turnover associated with rebalancing across numerous industry portfolios. AQR must assess whether the marginal alpha from industry momentum survives net of transaction costs and bid ask spreads. Additionally, the strategy's sensitivity to industry classification and signal definitions, such as the 12 to 2 month lookback, requires extensive robustness checks to ensure results are an enduring economic phenomenon rather than a product of backtest optimisation.

Ultimately, while the data supports the inclusion of industry momentum, its standalone performance has recently been statistically modest. AQR should proceed only if they can demonstrate that this three way combination delivers incremental, implementable value that cannot be replicated by simply increasing the tilt of existing multi factor products.
